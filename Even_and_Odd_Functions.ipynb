# Algebra with Python: Even and Odd Functions
## Introduction
In algebra, functions are classified as even or odd based on their symmetry. Understanding these concepts helps in graphing functions and simplifying expressions.

### What You'll Learn:
* Properties of even and odd functions
* How to determine whether a function is even or odd
* Implementing checks using Python

## Properties of Even and Odd Functions

### **Even Function**
A function $f(x) $ is even if:



$
f(-x) = f(x)
$



**Symmetry:** Even functions are symmetric about the **y-axis**.

**Examples:** 


$
f(x) = x^2, \quad f(x) = \cos(x)
$



### **Odd Function**
A function $f(x) $ is odd if:



$
f(-x) = -f(x)
$



**Symmetry:** Odd functions have rotational symmetry about the **origin**.

**Examples:** 


$
f(x) = x^3, \quad f(x) = \sin(x)
$




# Checking Even or Odd Functions with Python
## Example 1: Checking a Polynomial Function

```python
from sympy import symbols, simplify

x = symbols('x')

# Define function
def check_even_odd(f):
    neg_x_f = f.subs(x, -x)  # Replace x with -x
    if simplify(neg_x_f - f) == 0:
        return "Even Function"
    elif simplify(neg_x_f + f) == 0:
        return "Odd Function"
    else:
        return "Neither Even Nor Odd"

# Example function: x**2
f1 = x**2
print(check_even_odd(f1))  # Output: Even Function

# Example function: x**3
f2 = x**3
print(check_even_odd(f2))  # Output: Odd Function
```

# Example 2: Graphing Even and Odd Functions

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function(f, x_vals, title):
    y_vals = f(x_vals)
    plt.plot(x_vals, y_vals, label=title)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.legend()
    plt.grid()
    plt.show()

# Define even and odd functions
even_func = lambda x: x**2
odd_func = lambda x: x**3

x_vals = np.linspace(-10, 10, 400)

# Plot Even Function
plot_function(even_func, x_vals, "Even Function: f(x) = x^2")

# Plot Odd Function
plot_function(odd_func, x_vals, "Odd Function: f(x) = x^3")
```

## Summary

- **Even functions** satisfy $ f(-x) = f(x) $, showing symmetry about the **y-axis**.
- **Odd functions** satisfy $ f(-x) = -f(x) $, showing rotational symmetry about the **origin**.
- Python can be used to identify and visualize even and odd functions.
