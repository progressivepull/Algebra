# Advanced Algebra with Python: Exponential Functions

Exponential functions are mathematical expressions where the variable appears in the exponent, commonly written as:
$
f(x) = a \cdot b^x
$
where:
- $ a $ is the initial value,
- $ b $ is the base (growth or decay factor),
- $ x $ is the exponent (often time or another variable).

Exponential functions are widely used to model **growth** (e.g., populations, investments) and **decay** (e.g., radioactive decay, depreciation).

This tutorial shows how to analyze, plot, and apply exponential functions using Python with **NumPy**, **Matplotlib**, and **SymPy**.

## 1. Defining Exponential Functions

### Using NumPy and Matplotlib

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 2    # Initial value
b = 1.5  # Growth factor (b > 1 for growth, 0 < b < 1 for decay)

# Create x values
x = np.linspace(0, 10, 100)
y = a * b ** x

plt.plot(x, y, label="f(x) = 2 Ã— 1.5^x")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Exponential Function: Growth Example')
plt.grid(True)
plt.legend()
plt.show()
````

## 2. Modeling Exponential Decay

Set $ 0 < b < 1 $ for decay processes.

````python
a = 100
b = 0.8  # Decay factor

x = np.linspace(0, 20, 100)
y = a * b ** x

plt.plot(x, y, label="f(x) = 100 Ã— 0.8^x", color='red')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Exponential Function: Decay Example')
plt.grid(True)
plt.legend()
plt.show()
````

## 3. Exponential Functions with Base $ e $

The natural exponential function uses base $ e \approx 2.718 $:

$
f(x) = a \cdot e^{kx}
$

where $ k > 0 $ for growth and $ k < 0 $ for decay.

````python
a = 5
k = 0.3

x = np.linspace(0, 10, 100)
y = a * np.exp(k * x)

plt.plot(x, y, label="f(x) = 5 Ã— e^{0.3x}")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Natural Exponential Growth')
plt.grid(True)
plt.legend()
plt.show()
````

## 4. Symbolic Manipulation with SymPy

You can define and manipulate exponential functions symbolically.

````python
import sympy as sp

x = sp.symbols('x')
a, b = 2, 3

f = a * b ** x
print("Symbolic exponential function:", f)

# Differentiate
df = sp.diff(f, x)
print("Derivative:", df)

# Integrate
int_f = sp.integrate(f, x)
print("Integral:", int_f)
````

## 5. Applications: Modeling Growth and Decay

### Population Growth Example

$
P(t) = P_0 \cdot (1 + r)^t
$
Where:
- $ P_0 $ = initial population,
- $ r $ = growth rate per period,
- $ t $ = number of periods.

````python
P0 = 1000    # Initial population
r = 0.05     # 5% growth rate per year
t = np.arange(0, 21, 1)  # 0 to 20 years

P = P0 * (1 + r) ** t

plt.plot(t, P, marker='o')
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Exponential Population Growth')
plt.grid(True)
plt.show()
````

### Radioactive Decay Example

$
N(t) = N_0 \cdot e^{-kt}
$
Where:
- $ N_0 $ = initial amount,
- $ k $ = decay constant,
- $ t $ = time.

````python
N0 = 500
k = 0.2
t = np.linspace(0, 20, 100)
N = N0 * np.exp(-k * t)

plt.plot(t, N)
plt.xlabel('Time')
plt.ylabel('Amount Remaining')
plt.title('Radioactive Decay')
plt.grid(True)
plt.show()
````

## 6. Practice: Try Your Own Exponential Functions

Change the values of $ a $, $ b $, $ k $, and plot new functions or solve new symbolic problems.

````python
# Example: Compound interest
P0 = 2000
r = 0.03
n = 12  # Compounded monthly
t = np.linspace(0, 10, 100)
A = P0 * (1 + r/n) ** (n*t)

plt.plot(t, A)
plt.xlabel('Years')
plt.ylabel('Account Balance')
plt.title('Compound Interest Growth')
plt.grid(True)
plt.show()
````
