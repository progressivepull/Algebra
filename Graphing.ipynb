# Graphing and Visualizing Equations & Inequalities in Python

## Introduction

This tutorial will guide you through visualizing equations and inequalities on a coordinate plane using Python. We will use the popular libraries Matplotlib and NumPy to create graphs for:

- Linear equations
- Quadratic functions
- Other (custom) functions
- Inequalities (shaded regions)

## 1. Graphing Linear Equations

A linear equation has the form $ y = mx + b $.

**Example: $ y = 2x + 1 $**

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
y = 2 * x + 1

plt.plot(x, y, label='y = 2x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of Linear Equation')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
````

## 2. Graphing Quadratic Functions

A quadratic function has the form $ y = ax^2 + bx + c $.

**Example: $ y = x^2 - 4x + 3 $**

````python
x = np.linspace(-2, 6, 400)
y = x**2 - 4*x + 3

plt.plot(x, y, color='red', label='y = xÂ² - 4x + 3')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of Quadratic Function')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
````

## 3. Graphing Other Functions

You can graph any function by defining it in terms of `x`.

**Example: $ y = \sin(x) $ and $ y = \cos(x) $ on the same plot**

````python
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='y = sin(x)')
plt.plot(x, y2, label='y = cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphs of Sine and Cosine Functions')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
````

## 4. Visualizing Inequalities

To visualize inequalities like $ y \leq 2x + 1 $, shade the region that satisfies the inequality.

**Example: $ y \leq 2x + 1 $**

````python
x = np.linspace(-10, 10, 400)
y = 2 * x + 1

plt.plot(x, y, color='blue', label='y = 2x + 1')
plt.fill_between(x, y, -20, where=(y > -20), color='blue', alpha=0.2, label='y â¤ 2x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of Linear Inequality')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
````

**Example: $ y > x^2 - 4x + 3 $**

````python
x = np.linspace(-2, 6, 400)
y = x**2 - 4*x + 3

plt.plot(x, y, color='red', label='y = xÂ² - 4x + 3')
plt.fill_between(x, y, 10, where=(y < 10), color='red', alpha=0.2, label='y > xÂ² - 4x + 3')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of Quadratic Inequality')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
````

## 5. Combining Multiple Graphs

You can plot multiple equations and inequalities together for comparison.

**Example: Plot $ y = 2x + 1 $ and $ y = x^2 - 4x + 3 $ on the same graph**

````python
x = np.linspace(-2, 6, 400)
y1 = 2 * x + 1
y2 = x**2 - 4*x + 3

plt.plot(x, y1, label='y = 2x + 1')
plt.plot(x, y2, label='y = xÂ² - 4x + 3')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Multiple Functions')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
````
