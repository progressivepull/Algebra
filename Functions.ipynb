# Tutorial: Understanding and Plotting Functions in Python

Functions are fundamental mathematical objects that assign each input exactly one output. In algebra, functions are often written as \( f(x) \), where \( f \) is the function name and \( x \) is the input variable.

This tutorial demonstrates how to work with **functions** in Python, including **linear**, **quadratic**, **exponential**, and more. We’ll use the `numpy` and `matplotlib` libraries for calculations and plotting.

---

## 1. What is a Function?

A **function** is a rule that assigns to each input exactly one output.  
**Notation:** \( f(x) \), \( g(x) \), or \( y = f(x) \)

For example:  
- \( f(x) = 2x + 1 \) is a linear function.
- \( g(x) = x^2 \) is a quadratic function.

## 3. Defining Functions in Python

You can define a function using the `def` keyword:

````python
def linear(x):
    return 2*x + 1

print(linear(3))  # Output: 7
````

Or, use lambda functions for simple cases:

````python
quadratic = lambda x: x**2 - 4*x + 3
print(quadratic(2))  # Output: -1
````

## 4. Plotting Functions

Let’s visualize different types of functions.

````python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values
x = np.linspace(-10, 10, 400)

# Define functions
linear = lambda x: 2*x + 1
quadratic = lambda x: x**2 - 4*x + 3
exponential = lambda x: 2**x
cubic = lambda x: x**3 - 3*x**2 + x - 5

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, linear(x), label='Linear: $2x + 1$')
plt.plot(x, quadratic(x), label='Quadratic: $x^2 - 4x + 3$')
plt.plot(x, exponential(x), label='Exponential: $2^x$')
plt.plot(x, cubic(x), label='Cubic: $x^3 - 3x^2 + x - 5$')

plt.ylim(-10, 30)
plt.title("Different Types of Functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
````

## 5. Types of Functions

### **A. Linear Function**
Form: \( f(x) = mx + b \)
- Example: \( f(x) = 2x + 1 \)  
  *Graph is a straight line.*

### **B. Quadratic Function**
Form: \( f(x) = ax^2 + bx + c \)
- Example: \( f(x) = x^2 - 4x + 3 \)  
  *Graph is a parabola.*

### **C. Exponential Function**
Form: \( f(x) = a^x \)
- Example: \( f(x) = 2^x \)  
  *Graph grows rapidly for positive \( x \).*

### **D. Cubic and Others**
- **Cubic:** \( f(x) = x^3 - 3x^2 + x - 5 \)
- **Absolute Value:** \( f(x) = |x| \)
- **Logarithmic:** \( f(x) = \log(x) \) (for \( x > 0 \))

## 6. Practice: Try Defining Your Own!

Write Python functions for:

1. \( f(x) = |x| \) (absolute value)
2. \( f(x) = \log_2(x) \) (use `np.log2(x)` for \( x > 0 \))

Example:

````python
absolute_value = lambda x: np.abs(x)
logarithmic = lambda x: np.log2(x)

# Plot these as well!
x = np.linspace(-10, 10, 400)
plt.plot(x, absolute_value(x), label='Absolute: $|x|$')
plt.plot(x[x>0], logarithmic(x[x>0]), label='Logarithmic: $log_2(x)$')
plt.legend()
plt.grid(True)
plt.show()
````

## 7. Summary

- Define functions using `def` or `lambda`.
- Linear, quadratic, exponential, and other functions have unique forms and graphs.
- Use `numpy` for calculations and `matplotlib` for plotting.
