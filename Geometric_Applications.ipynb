## 2. Finding the Distance Between Two Points
The distance ùëë between two points $(x_1, y_1)$ and $(x_2, y_2)$ is given by:

$$d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$$

````python
import numpy as np

def distance(x1, y1, x2, y2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Example
print(distance(3, 4, 7, 1))  # Output: 5.0
````

## 3. Finding the Slope of a Line
The slope ùëö between two points $(x_1, y_1)$ and $(x_2, y_2)$ is given by:

$$m = \frac{y_2 - y_1}{x_2 - x_1}$$

````python
def slope(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)

# Example
print(slope(1, 2, 4, 8))  # Output: 2.0
````

## 4. Finding the Area of a Triangle
The area ùê¥ of a triangle given its three vertices $(x_1, y_1)$, $(x_2, y_2)$ and $(x_3, y_3)$ is:

$$A = \frac{\left| x_1(y_2 - y_3) + x_2(y_3 - y_1) + x_3(y_1 - y_2) \right|}{2}$$


````python
def triangle_area(x1, y1, x2, y2, x3, y3):
    return abs(x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)) / 2

# Example
print(triangle_area(0, 0, 5, 0, 0, 5))  # Output: 12.5
````

## 5. Visualizing Geometric Concepts
Let‚Äôs plot a triangle using **matplotlib** :

````python
import matplotlib.pyplot as plt

def plot_triangle(x1, y1, x2, y2, x3, y3):
    plt.plot([x1, x2, x3, x1], [y1, y2, y3, y1], 'bo-')
    plt.fill([x1, x2, x3], [y1, y2, y3], 'cyan', alpha=0.5)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Triangle Visualization")
    plt.grid(True)
    plt.show()

# Example
plot_triangle(0, 0, 5, 0, 0, 5)
````
