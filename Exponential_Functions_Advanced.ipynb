# Understanding Exponential Functions

An **exponential function** has the general form:

$
f(x) = a \cdot b^x
$

Where:

- $ a $ is the **initial value** (when $ x = 0 $)
- $ b $ is the **base** of the exponential function
- $ x $ is the **exponent**

### Growth and Decay Behavior
- If $ b > 1 $, the function models **exponential growth**.
- If $ 0 < b < 1 $, the function models **exponential decay**.

Exponential functions are commonly used in modeling populations, radioactive decay, compound interest, and various natural phenomena.


# Implementing Exponential Functions in Python

````python
import numpy as np
import matplotlib.pyplot as plt

# Define exponential function
def exponential_function(a, b, x):
    return a * (b ** x)

# Generate x values
x_values = np.linspace(-2, 2, 100)

# Compute y values for different bases
y_growth = exponential_function(1, 2, x_values)  # Growth (b=2)
y_decay = exponential_function(1, 0.5, x_values)  # Decay (b=0.5)

# Plot the functions
plt.figure(figsize=(8, 5))
plt.plot(x_values, y_growth, label="Growth (b=2)", color="blue")
plt.plot(x_values, y_decay, label="Decay (b=0.5)", color="red")
plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)
plt.legend()
plt.title("Exponential Growth vs Decay")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
````

# Example: Modeling Population Growth

Population growth can often be modeled using an **exponential growth function**:

$
P(t) = P_0 \cdot e^{rt}
$


Where:

- $ P_0 $ is the **initial population** (at $ t = 0 $)
- $ r $ is the **growth rate** (per unit of time)
- $ t $ is **time**
- $ e $ is **Euler’s number** ($\approx 2.718$)


````python
from scipy.optimize import curve_fit

# Define exponential growth model
def population_growth(t, P0, r):
    return P0 * np.exp(r * t)

# Generate time values
time = np.linspace(0, 10, 50)

# Compute population values for P0=100, r=0.2
population = population_growth(time, 100, 0.2)

# Plot the growth model
plt.figure(figsize=(8, 5))
plt.plot(time, population, label="Population Growth", color="green")
plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Population Growth Model")
plt.legend()
plt.grid(True)
plt.show()
````

# Example: Modeling Radioactive Decay

## Radioactive Decay Formula
The decay of a substance over time follows the exponential decay equation:

$
N(t) = N_0 \cdot e^{-kt}
$

Where:
- $ N_0 $ is the initial quantity of the substance
- $ k $ is the decay constant
- $ t $ is time
- $ e $ is Euler’s number, applied using `np.exp()`


````python
# Define decay function
def radioactive_decay(t, N0, k):
    return N0 * np.exp(-k * t)

# Generate time values
time = np.linspace(0, 10, 50)

# Compute decay values for N0=100, k=0.3
decay = radioactive_decay(time, 100, 0.3)

# Plot the decay model
plt.figure(figsize=(8, 5))
plt.plot(time, decay, label="Radioactive Decay", color="orange")
plt.xlabel("Time")
plt.ylabel("Remaining Quantity")
plt.title("Radioactive Decay Model")
plt.legend()
plt.grid(True)
plt.show()
````
