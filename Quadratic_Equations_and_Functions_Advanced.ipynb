## Quadratic Formula

The quadratic equation is of the form:

$ax^2 + bx + c = 0$

The quadratic formula is used to find the roots:

$
x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$

```python
import math

def quadratic_formula(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        return "No real roots"
    root1 = (-b + math.sqrt(discriminant)) / (2*a)
    root2 = (-b - math.sqrt(discriminant)) / (2*a)
    return root1, root2

# Example usage
print(quadratic_formula(1, -3, 2))  # Output: (2.0, 1.0)
```

## Graphing Quadratic Functions

A quadratic function of the form:



$
f(x) = ax^2 + bx + c
$



forms a **parabola**. The direction of the parabola depends on the coefficient $ a $:

- If $ a > 0 $, the parabola **opens upward**.
- If $ a < 0 $, the parabola **opens downward**.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define quadratic function
def f(x, a, b, c):
    return a*x**2 + b*x + c

# Generate x values
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals, 1, -3, 2)

# Plot function
plt.plot(x_vals, y_vals, label="f(x) = x^2 - 3x + 2")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel("x-axis")
plt.ylabel("f(x)")
plt.title("Graph of Quadratic Function")
plt.legend()
plt.grid()
plt.show()
```

## Analyzing Quadratic Functions

### **Key Properties**
- **Vertex**: The highest or lowest point of the parabola, located at:

$
x_{\text{vertex}} = \frac{-b}{2a}
$

- **Axis of Symmetry**: A vertical line that passes through the vertex:

$
x = \frac{-b}{2a}
$


- **Discriminant**: Determines the number of real roots of the quadratic equation $ ax^2 + bx + c = 0 $:

$
\text{Discriminant} = b^2 - 4ac
$


### **Python to Find the Vertex**

```python
def find_vertex(a, b):
    x_vertex = -b / (2 * a)
    return x_vertex

# Example usage
a, b = 1, -4
print(find_vertex(a, b))
```
