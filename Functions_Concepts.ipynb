## Function Notation

Mathematical notation for functions:

$ f(x) = x^2 + 3x - 5 $


```python
def f(x):
    return x**2 + 3*x - 5

# Example usage
print(f(2))  # Output: 5
```

# Domain and Range
* Domain: The set of all possible input values (x-values).
* Range: The set of possible output values (f(x)-values).

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function
def f(x):
    return x**2 + 3*x - 5

# Generate x values
x = np.linspace(-10, 10, 400)
y = f(x)

# Plot function
plt.plot(x, y)
plt.xlabel("Domain (x)")
plt.ylabel("Range (f(x))")
plt.title("Function: f(x) = x^2 + 3x - 5")
plt.grid()
plt.show()
```

## Domain and Range

### **Domain**
Since this is a quadratic function, it is defined for **all real numbers** ($ \mathbb{R} $):



$
\text{Domain}: (-\infty, \infty)
$



### **Range**
As the parabola opens **upward**, it has a minimum value at its vertex. The vertex occurs at:



$
x_{\text{vertex}} = \frac{-b}{2a} = \frac{-3}{2(1)} = -\frac{3}{2}
$



Evaluating $ f(-3/2) $:



$
f(-3/2) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5
$





$
= \frac{9}{4} - \frac{9}{2} - 5
$





$
= \frac{9}{4} - \frac{18}{4} - \frac{20}{4} = \frac{-29}{4}
$



Thus, the minimum value is $ -\frac{29}{4} $, meaning:



$
\text{Range}: \left[-\frac{29}{4}, \infty\right)
$




## Types of Functions

Different types of functions in algebra include:

### Linear Functions
$ f(x) = ax + b $

### Quadratic Functions
$ f(x) = ax^2 + bx + c $

### Polynomial Functions
$ f(x) = ax^n + \dots + bx + c $

### Exponential Functions
$ f(x) = a \cdot b^x $

```python
def linear(x, a, b):
    return a*x + b

def quadratic(x, a, b, c):
    return a*x**2 + b*x + c

print(linear(3, 2, 5))      # Output: 11
print(quadratic(3, 1, 2, 3)) # Output: 18
```
