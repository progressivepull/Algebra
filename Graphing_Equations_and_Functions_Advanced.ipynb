# 1. Plotting Linear Functions
A linear equation follows the form: $ y = mx + b $

Here's how to graph a simple linear function using Python:

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def linear_function(x):
    return 2*x + 3  # Example equation y = 2x + 3

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate y values
y = linear_function(x)

# Plot the function
plt.plot(x, y, label="y = 2x + 3")
plt.axhline(0, color='black', linewidth=0.5)  # X-axis
plt.axvline(0, color='black', linewidth=0.5)  # Y-axis
plt.legend()
plt.title("Graph of a Linear Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
````

# 2. Plotting Quadratic Functions
A quadratic equation follows the form: $ y = ax^2 + bx + c $

Example:

````python
def quadratic_function(x):
    return x**2 - 4*x + 2  # Example equation y = x² - 4x + 2

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate y values
y = quadratic_function(x)

# Plot the function
plt.plot(x, y, label="y = x² - 4x + 2", color='red')
plt.axhline(0, color='black', linewidth=0.5)  
plt.axvline(0, color='black', linewidth=0.5)  
plt.legend()
plt.title("Graph of a Quadratic Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
````

# 3. Plotting Polynomial and Trigonometric Functions
For more complex functions, such as cubic or trigonometric graphs:

````python
def cubic_function(x):
    return x**3 - 3*x**2 + x - 5  # Example equation

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate y values
y = cubic_function(x)

# Plot the function
plt.plot(x, y, label="y = x³ - 3x² + x - 5", color='green')
plt.axhline(0, color='black', linewidth=0.5)  
plt.axvline(0, color='black', linewidth=0.5)  
plt.legend()
plt.title("Graph of a Cubic Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
````

Similarly, you can graph trigonometric functions like sine and cosine:

````python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label="y = sin(x)", color='blue')
plt.plot(x, y_cos, label="y = cos(x)", color='orange')
plt.axhline(0, color='black', linewidth=0.5)  
plt.axvline(0, color='black', linewidth=0.5)  
plt.legend()
plt.title("Sine and Cosine Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
````
